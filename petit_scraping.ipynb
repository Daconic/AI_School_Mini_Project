# 페이지 수집 함수 작성🔥

```python
def get_one_page(page_no):
    '''
    한페이지 수집 함수
    '''
    url = f'https://petitions.assembly.go.kr/api/petits?pageIndex={page_no}&recordCountPerPage=8&sort=AGRE_END_DE-&searchCondition=sj&searchKeyword=&petitRealmCode=&sttusCode=PETIT_FORMATN,CMIT_FRWRD,PETIT_END&resultCode=BFE_OTHBC_WTHDRAW,PROGRS_WTHDRAW,PETIT_UNACPT,APPRVL_END_DSUSE,ETC_TRNSF&notInColumn=RESULT_CODE&beginDate=20210119&endDate=20230119&ageCd='
    response = requests.get(url)
    
    df = pd.read_json(response.text)
    try:
        edit_df = df[['petitSj', 'petitCn','petitRealmNm','jrsdCmitNm','resultCodeNm','agreCo','agreEndDe','petitId']]

        edit_df = edit_df.rename(columns={
        'petitId' : '청원ID', 
        'petitSj' : '제목', 
        'petitCn' : '내용', 
        'petitRealmNm' : '분야', 
        'jrsdCmitNm' : '담당부서', 
        'resultCodeNm' : '결과', 
        'agreEndDe' : '동의종료일', 
        'agreCo': '동의수'})
        return edit_df
    except Exception as e:
        return None
```

```python
page = get_one_page(57)
page
```

```python
# url에 불필요해 보이는 속성을 지웠다가 자꾸 목록에 안보이는 다른 청원들이 들어와서 일단 속성값 하나도 안지우고 페이지 넘버만 바꾸는 걸로 썼다....
```

```python
def get_pages():
    '''
    페이지 수집 반복 함수
    '''
    page_no = 1
    pages = []

    while True:
        page = get_one_page(page_no)
        if page is None:
            break
        pages.append(page)
        page_no += 1

    time.sleep(1)
    
    df = pd.concat(pages, ignore_index=True)
    
    return df
```

```python
all_petit =get_pages()
```
